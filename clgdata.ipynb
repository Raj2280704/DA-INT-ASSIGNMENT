```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('as_dataset_shuffled.csv')

# Assign realistic conversion rates for each program interest
conversion_rates = {
    'AI': 0.10,
    'Cloud Developer': 0.12,
    'Data Science': 0.08,
    'Design UI/UX': 0.15,
    'Electric Vehicle': 0.07,
    'Java Developer': 0.11,
    'Robotics': 0.09,
    'Software Developer': 0.13
}

# Map conversion rates to the dataframe
df['Conversion Rate'] = df['Program Interest'].map(conversion_rates)

# Calculate the total leads for each program
total_leads = df['Program Interest'].value_counts().rename_axis('Program Interest').reset_index(name='Total Leads')

# Merge total leads with the original dataframe
df = df.merge(total_leads, on='Program Interest')

# Calculate projected conversions
df['Projected Conversions'] = df['Total Leads'] * df['Conversion Rate']

# Group by program interest to get total leads and projected conversions
program_summary = df.groupby('Program Interest').agg({
    'Total Leads': 'first',
    'Projected Conversions': 'first'
})

# Calculate budget allocation based on projected conversions
total_budget = 100000  # Example total marketing budget
program_summary['Budget Allocation'] = (program_summary['Projected Conversions'] / program_summary['Projected Conversions'].sum()) * total_budget

print(program_summary)

# Save the summary to a CSV file
program_summary.to_csv('program_summary.csv')
```
